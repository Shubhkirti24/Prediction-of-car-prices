### using groupby function:
categorical varible has different categories:

```python
## to find the categories in a categorical variable:
df['drive-wheels'].unique()
```

```python
## creating a new dataframe for the groupby function from the existing dataframe:
df_group= df[['drive-wheels','body-style','price']]
df_group.head()
```

```python
## using groupby function:
df_group_test=df_group.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_group_test
```

##### creating a pivot table to better visualise the groupby function:
one variable displayed along the rows, one for the columns and the target varible fills up the table

```python
df_pivot= df_group_test.pivot(index='drive-wheels',columns='body-style')
df_pivot
```

```python
## using heatmaos to generate the pivot table:
plt.pcolor(df_pivot, cmap='RdBu')

plt.colorbar()
plt.show()
```

```python
## setting useful labels in the heat map:
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap='RdBu')

#label names
row_labels = df_pivot.columns.levels[1]
col_labels = df_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=45)

fig.colorbar(im)
plt.show()
```

## data correlation statistics and causaton:
pearson co-orelation, p_value, f_value

```python
from scipy import stats
```

```python
## we need to calculate the pearson correlation and the p values for:
# (wheelbase, horsepower, length, width, curb-weigth, engine-size, bore, cty-mpg and highway-mpg ) vs price independently:

#Between wheelbase and price:
pearson_coef, p_value = stats.pearsonr(df["wheel-base"],df["price"])
print ("The pearson coefficient value is {} and p_value is {}".format(pearson_coef,p_value))

## pearson value uses (pearsonr)
```

```python
#Between horsepower and price:
pearson_coef, p_value = stats.pearsonr(df["horsepower"],df["price"])
print ("The pearson coefficient value is {} and p_value is {}".format(pearson_coef,p_value))
```

```python
#Between length and price:
pearson_coef, p_value = stats.pearsonr(df["length"],df["price"])
print ("The pearson coefficient value is {} and p_value is {}".format(pearson_coef,p_value))
```

```python
#Between width and price:
pearson_coef, p_value = stats.pearsonr(df["width"],df["price"])
print ("The pearson coefficient value is {} and p_value is {}".format(pearson_coef,p_value))
```

## analysis of variance: ANOVA
ANOVA shows variance in the different categories of the same variable Groupby function is used:

```python
## first we use the groupby funtion to produce a df to see which variable we want to analyze:
### lets intially take drive wheels and see what impact it has on the target varible -price:

df_group_dw= df[['drive-wheels','price']].groupby(['drive-wheels'])
df_group_dw.head()
```

```python
## in order to get a particular group we can use the get_group method:
df_group_dw.get_group("rwd")["price"]
```

```python
## anova retunrn two values: f_test score(which shows the variation between the same group to variation of means in the group) and p_value which is the confidence interval:

## calculating anova:

anova_results= stats.f_oneway(df_group_dw.get_group("fwd")["price"], df_group_dw.get_group("4wd")["price"])
anova_results
```
