# Data wrangling-pre processing

```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
## using the given dataset from the url:
url= "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

## importing the dataset into a python dataframe:
auto= pd.read_csv(url,header=None)
```

```python
## checking the rows of the given dataframe:
auto.head()
```

```python
### you can also use the given method to directly put the headers : df = pd.read_csv(filename, names = headers), with names containg ht elist of the col names
```

```python
## giving new headers:
headers =["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peakrpm","city-mpg","highway-mpg","price"]

## putting these headers into the dataframe:
auto.columns= headers
```

```python
## checking the new dataframe with the new headers
auto.head(5)
```

#### Dealing with the missing values in the dataframe:

```python
## since the dataframe has "?" instead of NaN we need to replace the values with NaN(/np.nan converts the values):
import numpy as np
auto.replace("?", np.nan, inplace=True)

auto.head()
```

```python
## we need to identify the missing values in the dataframe now using isnull, or notnull functions:
number=auto.isnull()
```

```python
## we need to count the values of missing data using a for loop to check each variable in the dataset:
#### we are using counts, value_counts and values.tolist():
## .values returns thebdataframe as an array:
for column in number.columns.values.tolist():
    print (column)
    print (number[column].value_counts())
    print(" ")
```

```python
## categorizing the columns and choosing how to replace the columns:
# normalised- losses,stroke,bore, horesepower, peak-rpm with mean
# replace doors with 4 as most of the sedans are 4 doors (choose the .idxmax() method to calculate the maximum occurences in the sedan family)

# delete the missing rows as that is the variable we are actually trying to predict
```

```python
## calucalting the means of the respective columns:
# we need to convert the values into float first and then calculate the mean:

normal_mean= auto["normalized-losses"].astype("float").mean(axis=0)
print("normal mean :",normal_mean)

bore_mean= auto["bore"].astype("float").mean()
print("bore mean :",bore_mean)

stroke_mean= auto["stroke"].astype("float").mean()
print("stroke mean :",stroke_mean)

horsepower_mean= auto["horsepower"].astype("float").mean()
print("horesepower mean :",horsepower_mean)

peakrpm_mean = auto["peakrpm"].astype("float").mean()
print("peakrpm mean :",peakrpm_mean)

## replacing the nan values with themean values in the distribution:
auto["normalized-losses"].replace(np.nan,normal_mean, inplace=True)

auto["bore"].replace(np.nan,bore_mean, inplace=True)

auto["stroke"].replace(np.nan,stroke_mean, inplace=True)

auto["horsepower"].replace(np.nan,horsepower_mean, inplace=True)

auto["peakrpm"].replace(np.nan,peakrpm_mean, inplace=True)


## checking the values
auto.head(5)
```

```python
# replace doors with 4 as most of the sedans are 4 doors
auto["num-of-doors"].replace(np.nan,"four", inplace=True)
```

```python
## drop the rows that have empty prices:

auto.dropna(subset=["price"],axis=0, inplace= True)
```

```python
auto["price"].value_counts()
```

```python
## we need to reset the index because we dropped 2 rows:
auto.reset_index(drop=True, inplace= True)
```

```python
auto.index
```

## data standardization

```python
## creating a new column by converting the mpg into km/l
auto['city-mpg/100']= 235/auto['city-mpg']
```

```python
auto.head()
```

## data normalisation:

```python
## coverting a variable range to 0-1 so as to give equal weightage("length","width","height") :
auto['length']= auto['length']/auto['length'].max()
auto['width'] = auto['width']/auto['width'].max()
auto['height'] = auto['height']/auto['height'].max()
```

## binning the data: 
the process of tranforming continuous numerical variables into 'bins' of discrete categories for grouped analysis

```python
number=auto.isnull()

for column in number.columns.values.tolist():
    print (column)
    print (number[column].value_counts())
    print(" ")

    
```

```python
## using the price variable to be grouped into bins:
## copying the data in anothre dataframe in the correct type:

auto["price"]=auto["price"].astype("int",copy= True)
bins = np.linspace(min(auto["price"]), max(auto["price"]),4)
bins
```

```python
## creating a list of different intervals:
group_names=["low","medium","high"]
```

```python
## using cut from pandas to cut the intervals into different bins:

auto["price-binned"] = pd.cut(auto["price"], bins, labels= group_names, include_lowest= True)
```

```python
auto.head()
```

```python
## using histogram to plot the price variables:
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(auto["price"])

## setting lables:
plt.pyplot.xlabel("prices")
plt.pyplot.ylabel("count")
plt.pyplot.title("price bins")
```

```python
## using the bar plot to plot the categorical variables:
pyplot.bar(group_names, auto["price-binned"].value_counts())
```

## turning categorical variables into quantative variables for analysis: (indicator varibale)
depending on its occurance it is signed a (0,1) value to be analysed 


```python
## one hot encoding:
dummy_variables= pd.get_dummies(auto['fuel-type'])
```

```python
dummy_variables

## using concatenation to add htese two dataframes:
auto= pd.concat([auto, dummy_variables], axis=1)

## drop the fuel-type column:
auto.drop("fuel-type",axis=1, inplace=True)
```
